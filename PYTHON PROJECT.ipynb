```python
import yfinance as yf


tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

tesla_data.reset_index(inplace=True)

print(tesla_data.head())
```

```python
import yfinance as yf

# Download Tesla's financial data
tesla_financials = yf.Ticker("TSLA").financials

# Convert to DataFrame and transpose for better readability
tesla_revenue_df = tesla_financials.transpose()

# Display the last five rows
print(tesla_revenue_df.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset the index
tesla_data.reset_index(inplace=True)

# Plotting the 'Close' price over time
tesla_data.plot(x='Date', y='Close', title='Tesla Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```
